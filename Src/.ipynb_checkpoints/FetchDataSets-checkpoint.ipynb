```python
import os
import pandas as pd 
currentPath,_=os.getcwd().split("\\Src") 

def set_pandas_options() -> None:
    pd.options.display.max_columns = 1000
    pd.options.display.max_rows = 1000
    pd.options.display.max_colwidth = 199
    pd.options.display.width = None
    # pd.options.display.precision = 2  # set as needed
set_pandas_options()

#We load the three datasets
sourcePathDataSet1Star=currentPath + "\\Input\\one-star-michelin-restaurants.csv"
sourcePathDataSet2Star=currentPath + "\\Input\\two-stars-michelin-restaurants.csv"
sourcePathDataSet3Star=currentPath + "\\Input\\three-stars-michelin-restaurants.csv"
df1Star = pd.read_csv(sourcePathDataSet1Star)
df2Star = pd.read_csv(sourcePathDataSet2Star) 
df3Star = pd.read_csv(sourcePathDataSet3Star) 

#We introduce a new column for the number of stars in each dataframe
df1Star.insert(0, "stars", [1 for i in range(df1Star.shape[0])], True)
df2Star.insert(0, "stars", [2 for i in range(df2Star.shape[0])], True) 
df3Star.insert(0, "stars", [3 for i in range(df3Star.shape[0])], True) 

#We merge the three data frames by stacking them one on top of the other
df=pd.concat([df1Star,df2Star,df3Star]).reset_index(drop=True)
display(df)

df=df.drop(columns=["price","zipCode","year"])
display(df)

```

```python
#let's see which rows have null values.
display(df.isnull().sum())
display(df[df["city"].isnull()])

# We will fix the rows that have Nan values
df.at[152,'city']="Hong Kong"
df.at[166,'city']="Hong Kong"

display(df.isnull().sum())

#df.to_csv('out.csv')
display(df.head(10))
```

```python
# We are going to create a column for the country, copying it from the region, as replacing those regions which are not countries for the corresponding country
df.insert(6, "country", df["region"], True)
df["country"]=df["country"].replace("California", "United States")
df["country"]=df["country"].replace("Chicago", "United States")
df["country"]=df["country"].replace("New York City", "United States")
df["country"]=df["country"].replace("Washington DC", "United States")
df["country"]=df["country"].replace("Hong Kong", "China")
df["country"]=df["country"].replace("Macau", "China")
df["country"]=df["country"].replace("Rio de Janeiro", "Brasil")
df["country"]=df["country"].replace("Sao Paulo", "Brasil")
df["country"]=df["country"].replace("Taipei", "Taiwan")
display(df)    
```

```python
def pingUrl(url):
    import requests
    response = requests.get(url)
    return (response.status_code)
    
def scrapePrice(url):
    import requests
    from bs4 import BeautifulSoup
    response = requests.get(url)
    soup = BeautifulSoup(response.text.encode('utf-8'))
    table=soup.select('div[class="jumbotron__card-detail d-flex align-items-center"]')[0].select('li')[1]
    extractedText=table.text.replace(" ","").replace("\n","")
    if extractedText[-3:].isdigit()==False and "-" in extractedText:
        minPrice,maxPrice=extractedText[:-3].split("-")
        currency=extractedText[-3:]
    elif extractedText[-3:].isdigit()==False and "-" not in extractedText:
        minPrice=extractedText[:-3]
        maxPrice="N/A"
        currency=extractedText[-3:]
    elif extractedText[-3:].isdigit()==True and "-" in extractedText:
        minPrice,maxPrice=extractedText.split("-")
        currency="N/A"
    else:
        minPrice=extractedText
        maxPrice="N/A"
        currency="N/A"
    return ([minPrice, maxPrice, currency])

urlStatusL=[]
for urlElement in df["url"]:
    urlStatusL.append(pingUrl(urlElement))
print(urlStatusL) 


''' 
#def ScrapeExchangeRates():
'''
```

```python
#df.to_csv('dfprebadurls.csv')

df=pd.read_csv('dfprebadurls.csv')
#display(df)
print(df.shape)
print(df["url_status"]==200)
dfOnlyExistingRestaurants=df[df["url_status"] == 200]
print(dfOnlyExistingRestaurants.shape)

minPriceL=[]
maxPriceL=[]
currencyL=[]
for urlElement in dfOnlyExistingRestaurants["url"]:
    minPrice, maxPrice, currency = scrapePrice(urlElement)
    minPriceL.append(minPrice)
    maxPriceL.append(maxPrice)
    currencyL.append(currency)

print(minPriceL)
print(maxPriceL)
print(currencyL)
```
